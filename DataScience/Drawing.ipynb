```python
listA=[5,4,3,2,1]
listB=[1,2,3,5,4]

plt.plot(listA,listB)

plt.xlabel('listA')
plt.ylabel('listB')
plt.title("Plot MUST Graph")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,5,6,7,9]

plt.scatter(x,y, label='skitscat', color='k', s=25, marker ="o")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter')
plt.show()
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2], label="Example 1")
plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Example 2", color='g')
plt.legend()

plt.xlabel('bar number')
plt.ylabel('bar height')

plt.title('Example Chart')

plt.show()
```
